```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from scipy.optimize import curve_fit
```

```python
tab=pd.read_csv('tab.txt')
```

```python
bins1 = np.linspace(-0.3, 0.3, 16)
bins2 = np.linspace(-3, 3, 16)

bin_counts1, bin_edges1 = np.histogram(tab['c'], bins=bins1)
bin_counts2, bin_edges2 = np.histogram(tab['x1'], bins=bins2)

plt.hist(tab['c'], bins=bins1, alpha=0.5, color='g', edgecolor='black')

# Print the bin limits
for i in range(len(bin_edges1) - 1):
    bin_start = bin_edges1[i]
    bin_end = bin_edges1[i + 1]
    print(f"Bin {i+1} limits: {bin_start} - {bin_end}")

plt.xlabel('c')
plt.ylabel('Frequency')
plt.show()


plt.hist(tab['x1'], bins=bins2, alpha=0.5, color='g', edgecolor='black')

# Access and print the bin limits
for i in range(len(bin_edges1) - 1):
    bin_start = bin_edges2[i]
    bin_end = bin_edges2[i + 1]
    print(f"Bin {i+1} limits: {bin_start} - {bin_end}")

plt.xlabel('x1')
plt.ylabel('Frequency')
plt.show()
```
