Divide the clipped residuals in 5 groups and find the frequency of x1 and c for every group

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
def save_values_between_range(values: pd.Series, ranges, filenames):
    for range_index, (lower_limit, upper_limit) in enumerate(ranges):
        filtered_values = values[(values >= lower_limit) & (values <= upper_limit)]
        filtered_data = '\n'.join([f'{index},{value}' for index, value in filtered_values.items()])

        with open(filenames[range_index], 'w') as file:
            file.write(filtered_data)


data=pd.read_csv('./clipped_residuals.csv')
ranges = [(-0.6, -0.2), (-0.2, -0.04), (-0.04, 0.04), (0.04, 0.2), (0.2, 0.6)]
filenames = ['group1.csv', 'group2.csv', 'group3.csv', 'group4.csv', 'group5.csv']

save_values_between_range(data['b'], ranges, filenames)
```

```python
# Keep the clipped supernovae

newdt=pd.read_csv('./groups/group1/group1.csv')

old=pd.read_csv('./lists/salt3_nir_fixed/salt3_nir_fixed.txt', delim_whitespace=True)

tab1 = old[old.index.isin(newdt['0'])]


newdt=pd.read_csv('./groups/group2/group2.csv')

old=pd.read_csv('./lists/salt3_nir_fixed/salt3_nir_fixed.txt', delim_whitespace=True)

tab2 = old[old.index.isin(newdt['0'])]


newdt=pd.read_csv('./groups/group3/group3.csv')

old=pd.read_csv('./lists/salt3_nir_fixed/salt3_nir_fixed.txt', delim_whitespace=True)

tab3 = old[old.index.isin(newdt['0'])]


newdt=pd.read_csv('./groups/group4/group4.csv')

old=pd.read_csv('./lists/salt3_nir_fixed/salt3_nir_fixed.txt', delim_whitespace=True)

tab4 = old[old.index.isin(newdt['0'])]


newdt=pd.read_csv('./groups/group5/group5.csv')

old=pd.read_csv('./lists/salt3_nir_fixed/salt3_nir_fixed.txt', delim_whitespace=True)

tab5 = old[old.index.isin(newdt['0'])]
```

```python
bin_range1 = (-3.1, 3)

sns.histplot(tab1['x1'], color='blue', label='group1', kde=True, bins=20, alpha=0.2, binrange=bin_range1, edgecolor='none')
sns.histplot(tab2['x1'], color='red', label='group2', kde=True, bins=20, alpha=0.2, binrange=bin_range1, edgecolor='none')
sns.histplot(tab3['x1'], color='green', label='group3', kde=True, bins=20, alpha=0.2, binrange=bin_range1, edgecolor='none')
sns.histplot(tab4['x1'], color='purple', label='group4', kde=True, bins=20, alpha=0.2, binrange=bin_range1, edgecolor='none')
sns.histplot(tab5['x1'], color='orange', label='group5', kde=True, bins=20, alpha=0.2, binrange=bin_range1, edgecolor='none')

plt.legend()  
plt.xlabel('x1')  
plt.ylabel('Frequency')  
plt.show()  


bin_range2 = (-0.3, 0.31)

sns.histplot(tab1['c'], color='blue', label='group1', kde=True, bins=20, alpha=0.2, binrange=bin_range2, edgecolor='none')
sns.histplot(tab2['c'], color='red', label='group2', kde=True, bins=20, alpha=0.2, binrange=bin_range2, edgecolor='none')
sns.histplot(tab3['c'], color='green', label='group3', kde=True, bins=20, alpha=0.2, binrange=bin_range2, edgecolor='none')
sns.histplot(tab4['c'], color='purple', label='group4', kde=True, bins=20, alpha=0.2, binrange=bin_range2, edgecolor='none')
sns.histplot(tab5['c'], color='orange', label='group5', kde=True, bins=20, alpha=0.2, binrange=bin_range2, edgecolor='none')

plt.legend()  
plt.xlabel('c')  
plt.ylabel('Frequency')  
plt.show()  
```
