Compare the different models using their peak magnitudes

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import time
```

```python
tab=pd.read_csv('./models.csv')
```

```python
x = tab['salt3_nir_fixed']
y = tab['salt2_fixed']
```

```python
sns.jointplot(x=x, y=y, kind='reg')
plt.xlim(10,22)
plt.ylim(10,22)
plt.xlabel('salt3_nir_fixed')
plt.ylabel('salt2_fixed')

sns.jointplot(x=x, y=y, kind='scatter')
plt.xlim(10,22)
plt.ylim(10,22)
plt.xlabel('salt3_nir_fixed')
plt.ylabel('salt2_fixed')

sns.jointplot(x=x, y=y, kind = 'hex')
plt.xlim(10,22)
plt.ylim(10,22)
plt.xlabel('salt3_nir_fixed')
plt.ylabel('salt2_fixed')
```
